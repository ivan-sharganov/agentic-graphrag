# Содержание
### Строим граф для 1 сбщ
1. Стереть Neo4j базу до нуля (7688)
2. Вызов DeepSeek с промптом (cклеиваем строки в промпт (+ сбщ, + онто))
3. Валидация и исправление ABox JSON c помощью LLM
4. JSON -> Cypher конвертер (LLM возвращает JSON с узлами и нодами, чтобы в конт окно влезало. Cypher не влезал!)
5. Запись в файл cypher_checkpoints - слепки базы на каждое сбщ
6. Внесение инфы в базу (исполнение всех Cypher команд по очереди)

````python
def step_2_prompt_to_abox():
    import os
    import json
    import requests
    import time
    
    PROMPT_PATH = "prompt.txt"
    MSG_PATH = "message.txt"
    ONTO_PATH = "ontology.txt"
    
    DEEPSEEK_API_KEY = "sk-8da8031877dc4c39ae8d7b624c70f01f"
    if not DEEPSEEK_API_KEY:
        raise RuntimeError("Set env var DEEPSEEK_API_KEY first (export DEEPSEEK_API_KEY=...).")
    
    DEEPSEEK_BASE_URL = os.getenv("DEEPSEEK_BASE_URL", "https://api.deepseek.com").rstrip("/")
    MODEL =  "deepseek-reasoner" # "deepseek-chat"
    
    with open(PROMPT_PATH, "r", encoding="utf-8") as f:
        prompt_template = f.read()
    
    with open(MSG_PATH, "r", encoding="utf-8") as f:
        message_text = f.read().strip()
    
    with open(ONTO_PATH, "r", encoding="utf-8") as f:
        ontology_text = f.read().strip()
    
    prompt = (prompt_template
              .replace("{RDF_ONTOLOGY_HERE}", ontology_text)
              .replace("{TEXT_HERE}", message_text))
    
    url = f"{DEEPSEEK_BASE_URL}/v1/chat/completions"
    headers = {
        "Authorization": f"Bearer {DEEPSEEK_API_KEY}",
        "Content-Type": "application/json",
    }
    payload = {
        "model": MODEL,
        "messages": [
            {"role": "system", "content": "Return ONLY JSON. No markdown. Follow the template strictly."},
            {"role": "user", "content": prompt},
        ],
        "temperature": 0.0,
    }
    
    t0 = time.perf_counter()
    resp = requests.post(url, headers=headers, json=payload, timeout=180)
    elapsed = time.perf_counter() - t0
    print(f"DeepSeek API latency: {elapsed:.2f}s")
    resp.raise_for_status()
    data = resp.json()
    
    # DeepSeek API finish reason (transport-level)
    api_finish = data["choices"][0].get("finish_reason", "")
    content = data["choices"][0]["message"]["content"]
    
    # Save to variable for next cells
    answer = content
    
    emoji = "🟢" if api_finish != "length" else "🔴"
    print(f"{emoji} API finish_reason = {api_finish}")
    print(answer)  # preview
    return None
````

````python
# 4 JSON → Cypher генерация
````

````python
def step_4_abox_to_cypher():
    import json
    import re
    from typing import Any, Dict, List, Tuple
    
    # ==========================================
    # JSON ABox -> Cypher (canonical keys + EN-only)
    # ==========================================
    
    # ---------- Helpers ----------
    
    _CYRILLIC_RE = re.compile(r"[\u0400-\u04FF]")
    
    def contains_cyrillic(s: str) -> bool:
        return bool(_CYRILLIC_RE.search(s or ""))
    
    def strip_code_fences(s: str) -> str:
        s = (s or "").strip()
        if s.startswith("```"):
            s = re.sub(r"^```(?:json)?\s*", "", s)
            s = re.sub(r"\s*```$", "", s)
        return s.strip()
    
    def neo4j_local_name(name: str) -> str:
        """
        pcfr:entityName -> entityName
        full URI -> last fragment
        """
        if not name:
            return ""
        if ":" in name:
            return name.split(":")[-1]
        if "#" in name:
            return name.split("#")[-1]
        if "/" in name:
            return name.rsplit("/", 1)[-1]
        return name
    
    def neo4j_safe_ident(name: str) -> str:
        """
        Make a Neo4j-safe identifier for labels, rel-types, property keys.
        Keeps ASCII letters/digits/_ ; everything else -> _
        Also ensures it doesn't start with a digit.
        """
        n = neo4j_local_name(name)
        n = re.sub(r"[^A-Za-z0-9_]", "_", n)
        if not n:
            return "X"
        if n[0].isdigit():
            n = "_" + n
        return n
    
    def slugify_ascii(s: str) -> str:
        """
        Lowercase ASCII slug: letters/digits/_ only.
        Non-ASCII is dropped. Multiple _ collapsed.
        """
        s = (s or "").strip().lower()
        # Replace spaces and punctuation with underscores
        s = re.sub(r"[\s\-]+", "_", s)
        # Drop non-ascii
        s = s.encode("ascii", "ignore").decode("ascii")
        s = re.sub(r"[^a-z0-9_]", "_", s)
        s = re.sub(r"_+", "_", s).strip("_")
        return s or "x"
    
    # Canonical key prefixes per label
    KEY_PREFIX = {
        "City": "city",
        "Airport": "airport",
        "Country": "country",
        "Traveller": "traveller",
        "BorderOfficer": "officer",
        "Encounter": "encounter",
        "DocumentCheck": "doccheck",
        "Questioning": "questioning",
        "Stamping": "stamping",
        "Outcome": "outcome",
    }
    
    def canonical_key(label_raw: str, key_raw: Any, props: Dict[str, Any]) -> str:
        """
        Enforce ONE canonical key style to avoid duplicates like 'paris' vs 'city_paris'.
        Rule:
          - If key already starts with "<prefix>_" keep it (after ascii slug cleanup).
          - Else derive from pcfr:entityName (if present) or from key itself.
        """
        label = neo4j_local_name(label_raw)
        prefix = KEY_PREFIX.get(label, slugify_ascii(label))
    
        key_str = "" if key_raw is None else str(key_raw)
    
        # Prefer entityName slug when available (real-world nodes)
        ent_name = props.get("pcfr:entityName") or props.get("entityName") or props.get("pcfr:claimText")
        base = str(ent_name) if ent_name else key_str
    
        # Special-case airports: if entityName looks like IATA (3 letters), use it
        if label == "Airport" and ent_name:
            iata = str(ent_name).strip()
            if re.fullmatch(r"[A-Za-z]{3}", iata):
                base = iata
    
        base_slug = slugify_ascii(base)
    
        # If already canonical, keep, but normalize slug part
        m = re.fullmatch(r"([a-z0-9]+)_(.+)", slugify_ascii(key_str))
        if m and m.group(1) == prefix:
            return f"{prefix}_{slugify_ascii(m.group(2))}"
    
        return f"{prefix}_{base_slug}"
    
    def assert_english_only(value: str, where: str) -> None:
        if contains_cyrillic(value):
            raise ValueError(f"Cyrillic characters detected in {where}: {value!r}. "
                             f"Your LLM must output English/Latin-only strings.")
    
    def cypher_literal(v):
        if v is None:
            return "null"
        if isinstance(v, bool):
            return "true" if v else "false"
        if isinstance(v, (int, float)):
            return str(v)
    
        # lists/dicts: store as JSON string
        if isinstance(v, (list, dict)):
            s = json.dumps(v, ensure_ascii=True, separators=(",", ":"))
            s = s.replace("\\", "\\\\").replace("'", "\\'").replace("\n", "\\n").replace("\r", "\\r")
            return f"'{s}'"
    
        s = str(v)
        assert_english_only(s, "string literal")
        s = s.replace("\\", "\\\\").replace("'", "\\'").replace("\n", "\\n").replace("\r", "\\r")
        return f"'{s}'"
    
    # ---------- Main conversion ----------
    
    def abox_json_to_cypher(answer: str) -> List[str]:
        raw = strip_code_fences(answer)
        root = json.loads(raw)
    
        payload = root.get("payload", {}) or {}
        nodes = payload.get("nodes", []) or []
        rels = payload.get("relationships", []) or []
    
        # Normalize keys + validate EN-only strings in entityName/claimText/topic/etc.
        norm_nodes = []
        for node in nodes:
            label_raw = node.get("label", "")
            props = dict(node.get("properties", {}) or {})
            key_raw = node.get("key")
    
            if not label_raw or key_raw is None:
                continue
    
            # validate specific properties that must be EN-only
            for k in ("pcfr:entityName", "pcfr:claimText", "pcfr:topic", "pcfr:resultCode"):
                if k in props and isinstance(props[k], str):
                    assert_english_only(props[k], f"properties[{k}]")
    
            new_key = canonical_key(label_raw, key_raw, props)
            norm_nodes.append({"label": label_raw, "key": new_key, "properties": props})
    
        # Create a lookup from (label, old_key) to new_key so relationships stay consistent
        # We also map (label, canonical_key(label, old_key, props)) to itself.
        key_map: Dict[Tuple[str, str], str] = {}
        for node in nodes:
            label_raw = node.get("label", "")
            props = dict(node.get("properties", {}) or {})
            key_raw = node.get("key")
            if not label_raw or key_raw is None:
                continue
            new_key = canonical_key(label_raw, key_raw, props)
            key_map[(neo4j_local_name(label_raw), str(key_raw))] = new_key
            key_map[(neo4j_local_name(label_raw), new_key)] = new_key
    
        # ---------- Generate Constraints ----------
        labels = sorted({neo4j_safe_ident(n.get("label", "")) for n in norm_nodes if n.get("label")})
    
        cypher_statements: List[str] = []
        for label in labels:
            cypher_statements.append(
                f"CREATE CONSTRAINT IF NOT EXISTS FOR (n:{label}) REQUIRE n.key IS UNIQUE"
            )
    
        # ---------- Generate Node Statements ----------
        for node in norm_nodes:
            label_raw = node.get("label", "")
            key = node.get("key")
            props = node.get("properties", {}) or {}
    
            label = neo4j_safe_ident(label_raw)
    
            set_parts = []
            for k, v in props.items():
                prop_key = neo4j_safe_ident(k)
                # For safety: disallow cyrillic in any string property value
                if isinstance(v, str):
                    assert_english_only(v, f"properties[{k}]")
                set_parts.append(f"n.{prop_key}={cypher_literal(v)}")
    
            set_clause = (" SET " + ", ".join(set_parts)) if set_parts else ""
            stmt = f"MERGE (n:{label} {{key:{cypher_literal(key)}}}){set_clause}"
            cypher_statements.append(stmt)
    
        # ---------- Generate Relationship Statements ----------
        # Hard blocklist: project forbids Claim↔Encounter evidence links
        forbidden_rel_types = {"pcfr:supportedBy", "supportedBy", "pcfr:contradictedBy", "contradictedBy"}
    
        for rel in rels:
            f = rel.get("from", {}) or {}
            t = rel.get("to", {}) or {}
    
            fl_raw = f.get("label", "")
            fk = f.get("key")
            tl_raw = t.get("label", "")
            tk = t.get("key")
            rt_raw = rel.get("type", "")
    
            if not fl_raw or not tl_raw or not rt_raw or fk is None or tk is None:
                continue
    
            if rt_raw in forbidden_rel_types:
                continue
    
            fl_local = neo4j_local_name(fl_raw)
            tl_local = neo4j_local_name(tl_raw)
    
            fk_new = key_map.get((fl_local, str(fk)), str(fk))
            tk_new = key_map.get((tl_local, str(tk)), str(tk))
    
            fl = neo4j_safe_ident(fl_raw)
            tl = neo4j_safe_ident(tl_raw)
            rt = neo4j_safe_ident(rt_raw)
    
            stmt = (
                f"MATCH (a:{fl} {{key:{cypher_literal(fk_new)}}}), "
                f"(b:{tl} {{key:{cypher_literal(tk_new)}}}) "
                f"MERGE (a)-[:{rt}]->(b)"
            )
            cypher_statements.append(stmt)
    
        return cypher_statements
    
    
    
    cypher_statements = abox_json_to_cypher(answer)
    # for s in cypher_statements[:10]:
    #     print(s + ";")
    return None
````

````python
# 5 Сохранение Cypher в чекпоинт
````

````python
def step_5_save_cypher():
    import os
    import re
    
    ckpt_dir = "cypher_checkpoints"
    os.makedirs(ckpt_dir, exist_ok=True)
    
    existing = []
    for name in os.listdir(ckpt_dir):
        m = re.match(r"^cypher_(\d+)\.txt$", name)
        if m:
            existing.append(int(m.group(1)))
    
    next_num = max(existing, default=0) + 1
    out_path = os.path.join(ckpt_dir, f"cypher_{next_num}.txt")
    
    with open(out_path, "w", encoding="utf-8") as f:
        for stmt in cypher_statements:
            f.write(stmt)
            f.write(";")
    
    print(f"Saved {len(cypher_statements)} statements to {out_path}")
    return None
````

````python
# 6 исполнение Cypher (каждая строка автономна)
````

````python
def step_6_execute_cypher():
    from neo4j import GraphDatabase
    
    URI = "neo4j://localhost:7688"
    USER = "neo4j"
    PASSWORD = "12345678"
    DATABASE = "neo4j"
    
    driver = GraphDatabase.driver(URI, auth=(USER, PASSWORD))
    
    try:
        with driver.session(database=DATABASE) as session:
            for i, stmt in enumerate(cypher_statements, 1):
                session.run(stmt).consume()
        print(f"✅ Executed {len(cypher_statements)} statements.")
    finally:
        driver.close()    
    return None
````

````python
from __future__ import annotations
from pathlib import Path
from typing import Optional, Iterable
from neo4j import GraphDatabase


def _split_cypher_statements(text: str) -> list[str]:
    """
    Split on semicolons, ignoring empty chunks.
    (Assumes your generator doesn't put ';' inside string literals.)
    """
    parts = [p.strip() for p in text.split(";")]
    return [p for p in parts if p]


def run_cypher_file(
    cypher_path: str,
    uri: str = "neo4j://localhost:7688",
    user: str = "neo4j",
    password: str = "12345678",
    database: Optional[str] = None,
) -> None:
    """
    Reads Cypher statements from a file and executes them sequentially.
    Each statement is run as its own query (Neo4j requires 1 statement per run()).
    """
    path = Path(cypher_path)
    if not path.exists():
        raise FileNotFoundError(f"Cypher file not found: {path}")

    cypher_text = path.read_text(encoding="utf-8")
    statements = _split_cypher_statements(cypher_text)

    if not statements:
        print("No Cypher statements found in file.")
        return

    driver = GraphDatabase.driver(uri, auth=(user, password))
    try:
        with driver.session(database=database) as session:
            for i, stmt in enumerate(statements, start=1):
                session.run(stmt).consume()  # consume forces execution now
                # print(f"✅ {i}/{len(statements)} executed")
        print(f"✅ Executed {len(statements)} statements from {path.name}")
    finally:
        driver.close()

run_cypher_file("cypher_checkpoints/cypher_1.txt", uri="neo4j://localhost:7688", user="neo4j", password="12345678")
````
